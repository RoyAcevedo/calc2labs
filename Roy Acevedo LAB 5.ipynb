Lets start with a curve we know the length of, a semicircle. The equation of a semicircle with radius $r$ is $$\sqrt{1 - r^2}$$
We know that the length of this curve is $\pi r$, lets show this with a semicircle of radius 1.

```python
# Lets graph it first
xs = np.linspace(-1,1,100)
ys = np.sqrt(1 - xs**2)
plt.plot(xs, ys)
plt.grid(True)
plt.show()
```

### Arc Length

Now we need a formula to find the arc length on a smooth curve f(x) on the interval $[a,b]$ The arc length s is defined  as $$s=\int_{a}^{b}{\sqrt{1 + f'(x)}dx}$$
Lets use python to find this integral

```python
semicircle = sqrt(1 - x**2)
deriv = diff(semicircle, x)
arc = integrate(sqrt(1 + deriv**2), (x, -1, 1))
arc
```

### Surface area of revolutions

To find the surface area of revolution, we need to spin that arc length we found around an axis, similar to when we are finding volumes. Note how this formula is similar to those methods. We define $S$ as the surface area, and $r(x)$ as the distance between the graph $f$ and the axis of rotation. 
$$S = 2\pi\int_{a}^{b}{r(x)\sqrt{1 + f'(x)}}$$

* Can we say anything about $r(x)$ if we are revolving around $x=0$?
* What about if we revolve around $y=0$

```python
# Lets find the surface area of a sphere. To do this we need to revolve oursame semicircle around either axis.
# We need to be clever here. I am taking half of the arc, from 0 to 1 and doubleing it.
sphere = 2 * pi * integrate(x * sqrt(1 + deriv**2), (x, 0, 1))
2 * sphere
```
